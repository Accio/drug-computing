```python
from EnergyMin import *
%matplotlib inline
```

First we instantiate the `EnergyMin` class in the variable `minimization`:

```python
minimization = EnergyMin()
```

The function `add_well` takes arguments in the following order:

`add_well(x0, y0, width, depth)`

where (`x0`,`y0`) is the position of the center of the Gaussian well, and the `width` and `depth` parameters control how wide and deep the well is, respecitvely.

So, for instance, the first well added below will be placed at (45, 20) and will have a depth of 100 and a width of 150.

```python
minimization.add_well(50, 50, 100, 300)
minimization.add_well(25, 25, 50, 100)
minimization.add_well(35, 30, 75, 75)
```

**N.B.** there is another function in this class called `add_landscape_feature(function)` which takes one argument `function` which must be a string. This string is evaluated by the class, and applied to the numpy arrays which define the x-y plane (`xx` and `yy` respectively).

As an example of how to use this function, this is how you could define a landscape feature which is shaped like a bowl, with a minimum at the origin, and who's isosurfaces are concentric circles, getting higher as you move further away from the origin:

`minimization.add_landscape_feature("np.sqrt(np.power(xx, 2) + np.power(yy, 2))")`.

It is important to keep in mind that `xx` and `yy` are numpy arrays and must be treated as such. Were you to try to define the same landscape feature with this function: `(xx**2 + yy**2)**(1/2)`, it would not work properly.

Also, remember that the function argument to be added to the landscape must be entered as a string.

### Let's see what the landscape looks like:

```python
minimization.plot_landscape()
```

Now we'll add a ball to the landscape at (30, 42):

```python
minimization.add_ball(30, 42)
```
